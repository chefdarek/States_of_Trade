```
import io
prod = pd.read_excel(
    'USINDUSTRIALPRODUCTION.xls', 
    skiprows=14, )
```

```
prod.columns = ['Month', 'Primary Metals', 'Computer and Electronics',
                   'Automobiles and Parts', 'Furniture', 'Apparel']
```

```
prod.head()
```

```
prod_melt = pd.melt(prod, id_vars='Month', var_name='Industry', value_name='Index')
prod_melt.head(1)
```

```
plt.figure(dpi=1000)
plt.style.use(['dark_background'])
sns.lineplot(prod_melt.Month, prod_melt.Index, data=prod_melt, hue='Industry', legend='brief')
plt.title( "US Loses World's Apparel Production Leader Status 2008",
      fontsize=18, fontweight='bold')
#X LABEL AND Y LABEL
plt.ylabel('Index 2012=100, Monthly, Seasonally Adjusted', fontsize=8, fontweight='bold')
plt.xlabel('Years of Production', fontsize=16, fontweight='bold')
#ADD ANNOTATION LINE AS NEEDED TO LENGTH OF LIST AND CHANGE INDEX
plt.grid(b=True, which='major', axis='both', color='grey', linestyle='-', linewidth=.5)
plt.show();
```
